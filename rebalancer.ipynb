```python
from portfoliorebalancer import Rebalancer
```

# A Simple Portfolio Rebalancer


## 1. Define Portfolio

```python
asset_names = ['BTC', 'ETH', 'SOL']
target_asset_allocation = [0.6, 0.3, 0.1]
asset_positions = [0.14, 0.9, 5.0]
asset_prices = [44000.0, 3000.0, 100.0]
```

## 2. Rebalancing: Buy-only
Use additional funds to bring the portfolio as close as possible to the target asset allocation.

```python
funds = 1000.0
rebalancer_buy = Rebalancer(asset_names, asset_positions, asset_prices, target_asset_allocation, rebalancing_strategy='buy_only', funds=funds)
# hint: if you would like to see actual amounts, use normalize=False in the plot method
_, _ = rebalancer_buy.rebalance().plot('portfolio')
_, _ = rebalancer_buy.rebalance().plot('deviation')
_, _ = rebalancer_buy.plot('rebalanced_portfolio')
_, _ = rebalancer_buy.plot('rebalancing_amounts', normalized=False)
rebalancer_buy.portfolio_rebalanced[
    ['asset_positions', 'market_value', 'funds', 'rebalancing_amount', 'new_asset_positions', 'new_market_value',
     'new_total_market_value']]
```

## 3. Rebalancing: Buy-and-sell
Rebalance the portfolio via buying and selling of assets.

```python
rebalancer_buy_and_sell = Rebalancer(asset_names, asset_positions, asset_prices, target_asset_allocation)
_, _ = rebalancer_buy_and_sell.rebalance().plot('portfolio')
_, _ = rebalancer_buy_and_sell.rebalance().plot('deviation')
_, _ = rebalancer_buy_and_sell.plot('rebalanced_portfolio')
_, _ = rebalancer_buy_and_sell.plot('rebalancing_amounts', normalized=False)
rebalancer_buy_and_sell.portfolio_rebalanced[
    ['asset_positions', 'market_value', 'funds', 'rebalancing_amount', 'new_asset_positions', 'new_market_value',
     'new_total_market_value']]
```

## 4. Rebalancing: Sell-only
Remove funds to and bring the portfolio as close as possible to the target asset allocation.

```python
funds = -1000.0
rebalancer_sell = Rebalancer(asset_names, asset_positions, asset_prices, target_asset_allocation, rebalancing_strategy='sell_only', funds=funds)
# hint: if you would like to see actual amounts, use normalize=False in the plot method
_, _ = rebalancer_sell.rebalance().plot('portfolio')
_, _ = rebalancer_sell.rebalance().plot('deviation')
_, _ = rebalancer_sell.plot('rebalanced_portfolio')
_, _ = rebalancer_sell.plot('rebalancing_amounts', normalized=False)
rebalancer_sell.portfolio_rebalanced[
    ['asset_positions', 'market_value', 'funds', 'rebalancing_amount', 'new_asset_positions', 'new_market_value',
     'new_total_market_value']]
```
